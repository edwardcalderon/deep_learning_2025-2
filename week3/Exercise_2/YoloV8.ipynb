```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import numpy as np

# Step 1: Load the YOLOv8 model
model = YOLO('yolov8n.pt')

# Step 2: Run the model on an image (with car class only, class index 2)
results = model('carro5_internet.jpg', imgsz=1024, conf=0.8, iou=0.7, classes=[2, 3]) #

# Step 3: Get the result for the first image
predictions = results[0]

# Step 4: Manually filter the boxes based on confidence (e.g., only keep confidence > 0.7)
filtered_results = [det for det in predictions.boxes.data if det[4] > 0.3]  # Keep boxes with confidence > 0.7

# Step 5: Annotate the image with the filtered boxes
annotated_img = predictions.orig_img.copy()  # Make a copy of the original image

# Loop through filtered results and draw bounding boxes
for det in filtered_results:
    # Extract the bounding box coordinates and check for NaN
    x1, y1, x2, y2, conf, cls = det
    if not (np.isnan(x1) or np.isnan(y1) or np.isnan(x2) or np.isnan(y2)):
        # Convert coordinates to integers
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        # Draw the bounding box (color in BGR: Blue, Green, Red)
        cv2.rectangle(annotated_img, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
        # Add label text (class name and confidence score)
        label = f'Car {conf:.2f}'
        cv2.putText(annotated_img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Step 6: Display the annotated image using OpenCV
cv2.imshow('Filtered Results', annotated_img)
cv2.waitKey(0)  # Wait for a key press to close the image
cv2.destroyAllWindows()

# Step 7: Save the annotated image to a file
cv2.imwrite('filtered_annotated_image.jpg', annotated_img)
```
